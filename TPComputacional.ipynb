```
### TRABAJO PRÁCTICO COMPUTACIONAL
### PUNTO 2

import numpy as np
import math as mt
import matplotlib.pyplot as plt

### Valores dados en la consigna del TP

m = 1.0
g = 9.8
k = 60.0
R = 1.0
tita0 = np.pi/2
titaf = 2*np.pi

### Strings 
labelTita = 'Tita'
labelFv = 'Fv [N]'


### Función que devuelve la fuerza de vínculo según lo calculado en el Punto 2 del TP

def fuerza_vinculo (m, g, k, R, tita):
  return 3*m*g*np.cos(tita) + 2*k*R*mt.pow(np.sin(tita),2) - k*R
  
### Función que grafica y en función de x

def plot_y_vs_x (x, y, labelx, labely):
  plt.plot(x,y,'r,-')
  plt.xlabel(labelx)
  plt.ylabel(labely)

### Función que devuelve un array con los valores de y
### para cada valor de x. Requiere como precondición que se pase por
### parámetro un array con los valores de x y la dimensión del array

def build_array_y (xarr, m, g, k, R, dim, func):
  yarr = np.zeros(dim)
  for i in range(dim):
    yarr[i] = func(m,g,k,R,xarr[i])
  return yarr

### Función que genera un vector con distintos valores de x y
### hace uso de las funciones build_array_y(xarr, m, g, k, R, dim, func) y
### plot_y_vs_x(x, y, labelx, labely) para generar el respectivo array de valores
### de y y para graficar y vs x

def plot_yarr_vs_xarr (m,g,k,R,func,x0,xf,labelx,labely,gap):
  dim = mt.floor((xf-x0)/gap)
  xarr = np.linspace(x0, xf, dim)
  yarr = build_array_y(xarr,m,g,k,R,dim,func)
  plot_y_vs_x(xarr, yarr, labelx, labely)


### Implementación de la función plot_yarr_vs_xarr con los parámetros
### dados en el TP, la función para la fuerza de vínculo y un valor  
### harcodeado de sensibilidad de la curva que puede modificarse in situ

plot_yarr_vs_xarr (m,g,k,R,fuerza_vinculo,tita0,titaf,labelTita,labelFv,0.01)
```

```
### TRABAJO PRÁCTICO COMPUTACIONAL
### PUNTO 3


### Plot values
tita0 = 0
titaf = 2*np.pi
gap = 0.01

### Strings 
labelTitaDosPuntos = 'Tita_Dos_puntos [1/s^2]'


### Función que devuelve tita_dos_puntos según lo calculado en el Punto 3 del TP

def tita_dos_puntos (m, g, k, R, tita):
  return (g/R)*np.sin(tita) - (k/m)*np.sin(tita)*np.cos(tita)


### Implementación de la función plot_yarr_vs_xarr con los parámetros
### dados en el TP, la función para tita dos puntos y los valores de ploteo

plot_yarr_vs_xarr (m,g,k,R,tita_dos_puntos,tita0,titaf,labelTita,labelTitaDosPuntos,gap)

import matplotlib.pyplot as plt

```
